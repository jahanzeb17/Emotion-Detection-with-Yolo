```python
import supervision as sv
from ultralytics import YOLO

# Load the emotion detection model
model = YOLO("/content/emotion_detection_model.pt")

# Define the video processing function
def process_video(frame, _):
    # Run detection
    results = model(frame, imgsz=1280)[0]
    detections = sv.Detections.from_ultralytics(results)

    # Filter only by confidence (not by specific class)
    confidence_mask = detections.confidence > 0.5
    detections = detections[confidence_mask]

    # Annotators
    box_annotator = sv.BoxAnnotator(thickness=1)
    label_annotator = sv.LabelAnnotator(text_thickness=1, text_scale=1)

    # Generate labels for detected classes
    labels = [
        f"{model.names[class_id]} {confidence:.2f}"
        for _, _, confidence, class_id, _, _ in detections
    ]

    # Annotate the frame with bounding boxes and labels
    frame = box_annotator.annotate(scene=frame, detections=detections)
    frame = label_annotator.annotate(scene=frame, detections=detections, labels=labels)

    return frame

# Run inference and generate annotated video
sv.process_video(
    # input video
    source_path="/content/emotionVideo2.mp4",
    # output video
    target_path="emotionVideo3_output3.mp4",
    callback=process_video
)
```
